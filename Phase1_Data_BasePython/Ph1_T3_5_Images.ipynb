#### Loading image data

- First some libraries we'll need:

```python
import numpy as np
import matplotlib.pyplot as plt # mother of all visualization packages in Python
from PIL import Image, ImageOps # Pillow: extremely useful for image loading/editing
```

Can load image as matplotlib object:
- This has a ton of low level functionality
- But maybe too low level for simple operations.

```python
mpl_image = plt.imread("Data/dog.jpg")
fig, ax = plt.subplots()
ax.imshow(mpl_image)
```

#### PIL: Pillow
- Higher level image loading/manipulation.

Load image as PIL image object:

```python
image = Image.open("Data/dog.jpg")
type(image)
```

Displaying is really easy:

```python
display(image)
```

Let's rotate this doge:

```python
image.
```

```python
# Rotate 90 degrees
image_rotated = image.transpose(Image.ROTATE_90)
display(image_rotated)
```

Let's inspect this doge in grayscale:

```python
# Convert to grayscale
image_grayscale = ImageOps.grayscale(image)
display(image_grayscale)
```

Let's look at this doge's eye.

```python
# Crop (x1, y1, x2, y2)
image_cropped = image.crop((190, 90, 350, 145))
display(image_cropped)
```

Yes, he's judging you.

- You may need a reminder that the doge finds you inadequate.
- PIL Image has .save() method.

```python
image_cropped.save('Data/judgey_eyes.jpg')
```

PIL is a good way to:
- Quickly look at images.
- Quick and dirty image manipulation.
